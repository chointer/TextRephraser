```python
from RephraseText import Rephrase
from itertools import product
```

```python
rephrase = Rephrase()
```

```python
### Input
input_prompt = "I found a book in his dark room."
style_types = ["narrative", "emotional", "objective"]
length_types = ["compress", "maintain", "expand"]

outputs = {}

for (style, length) in product(style_types, length_types):
    outputs[(style, length)] = rephrase(input_prompt, style, length)
    
for (style, length) in product(style_types, length_types):
    print(style, length, "\t", outputs[(style, length)])
```
